```python
import matplotlib.pyplot as plt
import numpy as np
import pickle, json

with open('v6_batch5_0_0_pop_convergence_matrix.json', 'r') as f:
	data = json.load(f)
    
```

```python
data2 = []

for i,pre in enumerate(data['includePre']):
    for j,post in enumerate(data['includePre']):
        if data['connMatrix'][i][j]>0:
            print('%s %s %.2f' % (pre,post,data['connMatrix'][i][j]))
    
```

```python
figSize = (36,24)
fig = plt.figure(figsize=figSize)  # Open a new figure


plt.subplot(1,2,1)
plt.imshow(data['connMatrix'])
plt.colorbar()
plt.ylabel('Pre', fontsize=48)
plt.xlabel('Post', fontsize=48)
plt.xlim(7.5,13.5)
plt.ylim(-0.5,13.5)
plt.xticks(range(8,14), data['includePre'][8:14], fontsize=32, rotation=75);
plt.yticks(range(14), data['includePre'], fontsize=32, rotation=0);
plt.clim(5,30)


plt.subplot(1,2,2)
plt.imshow(data['connMatrix'])
plt.colorbar()
plt.xlabel('Post', fontsize=48)
plt.xlim(-0.5,7.5)
plt.ylim(-0.5,13.5)
plt.xticks(range(8), data['includePre'][0:8], fontsize=32, rotation=75);
plt.clim(0,300)


plt.savefig('convergence_S1_Th_13pops.png', facecolor = 'white' , dpi=300)
     
plt.show()
```
